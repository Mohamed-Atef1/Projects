```python
import numpy as np
import requests 
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
urls = "https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analysis&start="
df_list = []

for i in range(0,2):
    url = urls+str(i)

    response = requests.get(url)

    html = response.content

    soup = bs(html , "html.parser")

    containers = soup.find_all('div' , class_=('css-1gatmva e1v1l3u10'))

    for container in containers :
        j_title = container.find('h2' , class_=('css-m604qf')).text.strip()
        c_name = container.find('a' , class_=('css-17s97q8')).text.strip()
        c_land = container.find('span' , class_=('css-5wys0k')).text.strip()
        j_dis = container.find('span' , class_=('css-1ve4b75 eoyjyou0')).text.strip()


        df_list.append({'j_title' : j_title , 'c_name' : c_name , 'c_land' : c_land , 'j_dis' : j_dis})
        
```

```python
df = pd.DataFrame(df_list , columns=['j_title' , 'c_name' , 'c_land' , 'j_dis'])
```
